# Bayesian parameter estimation

- We will do Bayesian parameter estimation to infer distances to field stars from parallaxes using a Galaxy distance prior
- We will use topcat to get parallaxes for a list of stars
- We will infer the distance to a cluster using the parallaxes of its member stars

Resources:
- [GDR2mock](https://arxiv.org/abs/1804.01427)
- [Geometric distances paper](https://arxiv.org/abs/1804.10121)

- [Coryn's cluster distances tutorial](https://github.com/ehalley/parallax-tutorial-2018/blob/master/Distance%20inference%20-%20multiple%20sources.ipynb)

### Posterior: P(d|&varpi;,&sigma;<sub>&varpi;</sub>) = (Likelihood * Prior) / Evidence
### Likelihood: P(&varpi;|d,&sigma;<sub>&varpi;</sub>)
### Prior: P(d)
### Evidence: Integral of the Posterior over the parameterspace. We can neglect that for Parameter estimation.

````python
def get_percentiles(function,grid,percentiles=[0.14,0.5,0.86]):
    '''
    function evluated on the grid values
    
    returns the grid points where the cumulative function increases above a certain percentile 
    '''
    function /= sum(function)
    values = []
    percentile_counter = 0
    for i,item in enumerate(np.cumsum(function)):
        if item > percentiles[percentile_counter]:
            values.append(grid[i])
            percentile_counter += 1
            if percentile_counter == len(percentiles):
                break        
    return(values)

def likelihood_function(distance,parallax,uncertainty):
    '''
    distance in pc
    parallax in mas
    uncertainty in mas
    
    returns unnormalized likelihood function
    '''
    parallax /= 1000
    uncertainty /= 1000
    result = np.exp(-np.divide((parallax-np.divide(1,distance))**2,2*uncertainty**2))
    return(result)

# we do the posterior calculation for 3 different parallax uncertainties
for sigma_varpi in [0.5,1.5,4.5]:
    measured_parallax = 10
    print('uncertainty: ', sigma_varpi, 'mas, fractional parallax uncertainty: ', sigma_varpi/measured_parallax)
    print('##############################################################################################')
    dist_range = np.linspace(0,400,4001)

    likelihood = likelihood_function(dist_range,measured_parallax, sigma_varpi)
    plt.plot(dist_range,likelihood)
    plt.xlabel('likelihood function [pc]')
    plt.show()
    plt.clf()
    
    print("likelihood 14,50,84 percentiles in pc: ",get_percentiles(likelihood,dist_range))
    
    prior_dist = np.interp(dist_range,prior.dist,prior.ct,left = 0, right = 0)
    posterior = likelihood * prior_dist

    plt.plot(dist_range,posterior)
    plt.xlabel('posterior distance [pc]')
    plt.show()
    plt.clf()
    print("posterior 14,50,84 percentiles in pc: %s",str(get_percentiles(posterior,dist_range)))
    print('---------------------------------------------------------------------------------------------')
    print(' ')
plt.close()
````

````python
# We will simply increase the internal/reported error according to the values 
# for faint stars mentioned in the presentation slide 17
def external_uncertainty(internal_uncertainty):
    k = 1.08
    sigma_s = 0.043
    return(np.sqrt( k**2 * internal_uncertainty**2 + sigma_s**2))
print(np.median(data.parallax_error))
data.parallax_error = external_uncertainty(np.hstack(data.parallax_error))
print(np.median(data.parallax_error))
````

## Naive parallax combination (everything below is adapted from Coryn Bailer-Jones parallax tutorial)

The following methods for estimating the cluster distance from some combination of the parallaxes and their uncertainties are poor.

First, it is quite wrong to estimate the cluster distance via the mean of the inverse parallaxes (image some parallaxes were negative), and is meaningless when the fractional uncertainties are large. I do it here for comparison (and it's a fair way of getting a feel for the order-of-magnitude that the data imply).

````python
# Inverse parallaxes tell us
distances = np.divide(1,data.parallax)
print('mean and std of the inversed parallaxes: %.3f +- %.3f kpc'%(np.mean(distances), np.std(distances)))
````

The next thing we could think about doing is taking the inverse of the mean of parallaxes, and estimating its variance by a first order Taylor expansion. This is a slight improvement, but still not what we should be doing.

````python
# The inverse of the mean of the parallaxes:
Nstar = len(data)
mpar = np.mean(data.parallax)
rNaive1 = np.divide(1,mpar)
rNaive1SD = (1/Nstar)*np.sqrt(sum(data.parallax_error**2)) / mpar**2
print("Inverse of mean parallaxes: ", rNaive1, "+/-", rNaive1SD, "kpc\n")
# This gives an estimate of the distance to the cluster, together with the standard deviation in this mean.
print("SD of inverse mean parallax in the sense of a spread:", np.std(data.parallax_error)/mpar**2, "kpc\n")

````

We could go one step further and take a (variance) weighted mean. Specifically, take the inverse of the variance weighted mean of the parallaxes and estimate the variance of this again by a first order Taylor expansion.

````python
wfac = np.sum(1/data.parallax_error**2)
wmpar = np.sum(np.divide(data.parallax,data.parallax_error**2))/wfac
rNaive2 = 1/wmpar
rNaive2SD = 1/(wmpar**2 * np.sqrt(wfac))
print("Inverse of weighted mean of parallaxes: ", rNaive2, "+/-", rNaive2SD, "kpc\n")
print("SD of inverse weighted mean parallax in the sense of a spread:",
      np.std(data.parallax_error)/wmpar**2, "kpc\n")

````

# Estimate distance to cluster (no correlations or cluster spread)

Rather than mucking about with inverting and averaging the parallaxes, we should think about what we want to achieve. We want to infer the distance, $r$, to the cluster given a set of parallaxes $\{\varpi\}$ and corresponding parallax uncertainties $\{\sigma_{\varpi}\}$ of (some of) its members. Let's assume, for now, that the cluster is very small compared to its distance from us, such that we can consider the true distance to all the stars to be the same ("no cluster spread"), and equal to the distance of cluster, $r$. We proceed as for the single star case, but now using a multivariate likelihood. For $N$ stars this is an $N$-dimensional Gaussian. If we assume (for now) that the parallaxes are independent ("no correlations"), then this is just a product of $N$ 1-dimensional Gaussians. The posterior is

\begin{alignat}{2}
P(r|\{\varpi\}, \{\sigma_{\varpi}\}) 
&\propto \ P(r) \ P(\{\varpi\} | r, \{\sigma_{\varpi}\}) \\
&\propto \ P(r)  \prod_i P(\varpi_i | r, \sigma_{\varpi_i})
\end{alignat}

# Estimate distance to cluster (with correlations, no cluster spread)
## Invent a simple model for spatial parallax correlations
I now allow for correlations between the parallaxes. Given a set of $N$ parallax measurements, their correlations are determined by the $N \times N$ covariance matrix $C$. The likelihood for the $N$ measurements is no longer a product of $N$ one-dimensional Gaussian functions, but is rather an $N$-dimensional Gaussian

$
P({\boldsymbol \varpi} | {\mathbf r}, C) = \frac{1}{(2\pi)^{N/2}|C|^{1/2}}\exp\left[-\frac{1}{2}\left({\boldsymbol \varpi} - \frac{1}{{\mathbf r}}\right)^\mathsf{T} C^{-1} \left({\boldsymbol \varpi} - \frac{1}{{\mathbf r}}\right)\right]
$

where

$
\frac{1}{{\mathbf r}} = \left(\frac{1}{r_1}, \frac{1}{r_2}, \ldots \right) .
$

To illustrate this approach I invent a model (unrelated to Gaia, but do check [astrometric solution paper figure 14](https://www.aanda.org/articles/aa/abs/2018/08/aa32727-18/aa32727-18.html)) which specifies that the correlation between any two parallax measurements, $\varpi_1$ and $\varpi_2$, is a function of the angular separation, $\theta$, of the two sources on the sky

$
\rho(\varpi_1, \varpi_2) = a\exp(-\theta/\theta_{\rm len})
$


where $\theta_{\rm len}>0$ is some fixed angular scale and $0\leq a \leq 1$ gives the amplitude of the correlation at zero separation. The elements of the covariance matrix $C$ are 

$
C_{i,j} = \rho(\varpi_i, \varpi_j)\sigma_{\varpi_i}\sigma_{\varpi_j}.
$

Provided the two sources are not widely separated and not near the poles, we can approximate their angular separation as

$
\theta = \left( [(\alpha_1\!-\!\alpha_2)\cos(0.5(\delta_1\!+\!\delta_2))]^2 + (\delta_1\!-\!\delta_2)^2  \right)^{1/2}
$

where $(\alpha, \delta)$ are the RA and declination of a source.

In the code that follows, the function 'correlation' computes $\rho$, and nested for loop is calculating $C$.

````python
# The following code is still under construction
# define and plot the correlation model
# Play around with those values and see what the impact is
a = 0.5
theta_len = 0.1


# function that calculates the 
def ang_sep(ra1,ra2,dec1,dec2):
    return(np.sqrt(( (ra1-ra2) * np.cos(np.deg2rad(0.5*(dec1+dec2))) )**2 + (dec1 - dec2)**2))

def correlation(a,theta_len,separation):
    return(a*np.exp(np.divide(-separation,theta_len)))

angles = np.linspace(0,1,51)

plt.plot(angles,correlation(a,theta_len,angles))
plt.xlabel('angular separation [deg]')
plt.ylabel(r'correlation of $\sigma_\varpi$')
plt.show()
plt.clf()
````
